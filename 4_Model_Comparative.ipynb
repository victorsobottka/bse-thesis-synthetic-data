```python
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.heatmap(
    df[["K-S $p$-value", "Welch $p$-value"]],
    annot=True, cmap='viridis', cbar_kws={'label': '$p$-value'},
    yticklabels=models, vmin=0, vmax=1
)
plt.title("Heatmap of $p$-values for Statistical Tests")
plt.ylabel("Model")
plt.show()
```
